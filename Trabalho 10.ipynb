````python
def concatena_horizontalmente(A, B, C, separador):
    A = open(A, "r")
    B = open(B, "r")
    C = open(C, "w")
    tamA, tamB = len(A.readlines()), len(B.readlines())
    A.seek(0)
    B.seek(0)
    if tamA > tamB:
      for i in range(tamB):
        C.write(A.readline().strip("\n") + separador + B.readline().strip("\n") + "\n")
      for item in A:
        C.write(item)
    else:
      for i in range(tamA):
        C.write(A.readline().strip("\n") + separador + B.readline().strip("\n") + "\n")
      for item in B:
        C.write(item)

    A.close()
    B.close()
    C.close()
````

````python
def quadro_de_medalhas(arquivo='medalhas.txt'):
  arq = open(arquivo)
  quadro = open("quadro_de_medalhas.txt", "w")
  paises = {'BRA': {'bronze': 0,'prata': 0,'ouro': 0}, 'EUA': {'bronze': 0,'prata': 0,'ouro': 0}, 'CHN': {'bronze': 0,'prata': 0,'ouro': 0}, 'JPN': {'bronze': 0,'prata': 0,'ouro': 0}, 'RUS': {'bronze': 0,'prata': 0,'ouro': 0}, 'ITA': {'bronze': 0,'prata': 0,'ouro': 0}, 'GER': {'bronze': 0,'prata': 0,'ouro': 0}, 'GBR': {'bronze': 0,'prata': 0,'ouro': 0}, 'FRA': {'bronze': 0,'prata': 0,'ouro': 0}, 'CUB': {'bronze': 0,'prata': 0,'ouro': 0}, 'CAN': {'bronze': 0,'prata': 0,'ouro': 0}, 'AUS': {'bronze': 0,'prata': 0,'ouro': 0}, 'ARG': {'bronze': 0,'prata': 0,'ouro': 0}}

  medals = [medal.split() for medal in arq.readlines()]

  for medal in medals:
    paises[medal[0]][medal[1]] += 1
    
  results = sorted(dict(sorted(dict(sorted(dict(sorted(paises.items(), key=lambda item: item[0], reverse=True)).items(), key=lambda item: item[1]['bronze'], reverse=True)).items(), key=lambda item: item[1]['prata'], reverse=True)).items(), key=lambda item: item[1]['ouro'], reverse=True)

  quadro.write('pos\tpaís\tO\tP\tB\n')
  cont = 0
  for result in results:
    cont+=1
    quadro.write(str(cont)+'\t'+str(result[0])+'\t'+str(result[1]['ouro'])+'\t'+str(result[1]['prata'])+'\t'+str(result[1]['bronze']) + '\n')

  arq.close()
  quadro.close()
````
