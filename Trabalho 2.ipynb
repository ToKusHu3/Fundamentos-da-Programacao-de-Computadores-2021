
##Exercício 1

Faça um programa que calcule as raízes reais de uma equação do segundo grau f(x) = ax^2 + bx + cf(x)=ax 
2
 +bx+c usando a fórmula de Bhaskara.

Entrada:
```
a b c
```
Saída:
No caso de duas raízes reais:
```
x1
x2
```
Ou, no caso de apenas uma raiz real:
```
x1
```
Ou, no caso de não existirem raízes reais, a mensagem:
```
No real roots.
```


````python
from math import sqrt
a, b, c = [float(num) for num in input().split()]
delta = b**2 - 4 * a * c
if delta == 0:
  print("%.2f" % ((-b) / (2 * a)))
elif delta < 0:
  print("No real roots.")
elif delta > 0:
  print("%.2f" % ((-b + sqrt(delta)) / (2 * a)))
  print("%.2f" % ((-b - sqrt(delta)) / (2 * a)))
````

##Exercício 3
Faça um programa que leia três coordenadas em R²
  e indique se formam um triângulo, juntamente com o seu tipo: equilátero, isósceles ou escaleno.

Entrada:
```
x1 y1
x2 y2
x3 y3
```
Saída:
```
triângulo/não triângulo
equilátero/isósceles/escaleno
```

````python
def areaTriangulo(p1, p2, p3):
  return p1[0] * p2[1] + p2[0] * p3[1] + p3[0] * p1[1] - (p1[1] * p2[0] + p2[1] * p3[0] + p3[1] * p1[0])

def distPr2(p1, p2):
  return ((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2) ** (1/2)

v1 = [float(num) for num in input().split()]
v2 = [float(num) for num in input().split()]
v3 = [float(num) for num in input().split()]
area = areaTriangulo(v1, v2, v3)
l1 = distPr2(v1, v2)
l2 = distPr2(v2, v3)
l3 = distPr2(v3, v1)

if area == 0:
  print("não triângulo")
else:
  print("triângulo")
  if l1 == l2 == l3:
    print("equilátero")
  elif l1 == l2 or l1 == l3 or l2 == l3:
    print("isóceles")
  else:
    print("escaleno")

````

##Exercício 8

Dois times, Corinthians e Flamengo, participam de um campeonato de futebol, ao lado de outros times. Cada vitória confere três pontos e cada empate, um ponto. As equipes que ganham mais pontos são as melhores classificadas no campeonato. Em caso de empate no número de pontos, o time com maior saldo de gols é o melhor colocado. Se o número de pontos e a diferença de gols forem iguais para as duas equipes, as duas equipes estão empatadas.

Dado o número de vitórias e empates, e a quantidade de gols das duas equipes, sua tarefa é determinar qual das duas equipes está melhor classificada, ou se estão empatadas no campeonato.

Entrada:
A entrada é descrita em duas linhas, que contém três inteiros cada, separados por um espaço em branco
```
Cv Ce Cs
Fv Fe Fs
```
Na primeira linha temos o número de vitórias (Cv), o número de empates (Ce) e o saldo de gols (Cs) do Corinthians, respectivamente. Já na segunda linha temos o número de vitórias (Fv), o número de empates (Fe) e o saldo de gols (Fs) do Flamengo, respectivamente.

Saída:
Seu programa deve imprimir uma única linha. 

Caso o Corinthians esteja melhor classificado:
```
Corinthians 
```
Caso o Flamengo esteja melhor classificado:
```
Flamengo 
```
Ou, caso os times estejam empatados:
``` 
Empate
```

````python
Cp = [int(num) for num in input().split()]
Fp = [int(num) for num in input().split()]

Cpts = 3*Cp[0] + Cp[1]
Fpts = 3*Fp[0] + Fp[1]

if Cpts > Fpts:
  print("Corinthians")
elif Cpts < Fpts:
  print("Flamengo")
else:
  if Cp[2] > Fp[2]:
    print("Corinthians")
  elif Cp[2] < Fp[2]:
    print("Flamengo")
  else:
    print("Empate")
````

Na roleta, os bolsos são numerados de 0 a 36. Abaixo estão as cores dos bolsos:



*   bolso 0 verde;
*   para os bolsos de 1 a 10, os bolsos ímpares são vermelhos e os bolsos pares são pretos;
*   para os bolsos 11 a 18, os bolsos ímpares são pretos e os bolsos pares são vermelhos;
*   para os bolsos de 19 a 28, os bolsos ímpares são vermelhos e os bolsos pares são pretos;
*   para os bolsos de 29 a 36, ​​os bolsos ímpares são pretos e os pares, vermelhos.




Escreva um programa que leia o número do bolso e indique se o bolso é verde, vermelho ou preto. O programa deve exibir uma mensagem de erro se o usuário inserir um número fora do intervalo de 0 a 36.

````python
n = int(input())
if 0 <= n <= 36:
  if n == 0:
    print("verde")
  elif 1 <= n <= 10 or 19 <= n <= 28:
    if n % 2 == 0:
      print("preto")
    else:
      print("vermelho")
  else:
    if n % 2 == 0:
      print("vermelho")
    else:
      print("preto")
else:
  print("entrada inválida")
````
