## Problema 2
Faça um programa que dado o limite superior do somatório nn , calcule uma aproximação de π usando a fórmula Bailey-Borwein-Plouffe (BBP).

Neste problema em específico, use 48 casas decimais.

Note que o somatório é definido para k=0,…,∞. Para uma aproximação com um número fixo de termos ele passa a ser k=0,…,n.
Como os termos são valores muito próximos da precisão de armazenamento de números reais em um computador, a ordem da soma pode interferir no resultado devido à propagação do erro de representação ao somar cada termo.
Para evitar erros desse tipo, siga a ordem de k=0,…,n. Evite somar os termos na ordem inversa (k=n,…,0).Captura de tela 2021-09-30 190358.jpg

````python
n = int(input())
soma = 0
for i in range(n+1):
  soma += (1 / 16 ** i) * ((4 / (8 * i + 1)) - (2 / (8 * i + 4)) - (1 / (8 * i + 5)) - (1 / (8 * i + 6)))
print("%.48f" % soma)
````

##Problema 4
Faça um programa que imprima os horários de um ônibus, sabendo que ele passa a cada t minutos e que as linhas começam a funcionar às 5:00 (primeiro horário) e param às 23:55 (último horário possível).

Para imprimir as horas e minutos sempre com 2 dígitos, incluindo zeros à esquerda, utilize o comando:

print('%02d:%02d' % (horas, minutos))

Entrada
```
t
```
Saída
```
HH:MM
HH:MM
...
HH:MM
```

````python
def formataHoras(min):
  hrs = min//60
  minuts = min%60
  return ("%02d:%02d" % (hrs, minuts))

n = int(input())
for i in range(300, 1436, n):
  print(formataHoras(i))
  
````

##Problema 5
Faça um programa que verifica se um número inteiro positivo 0 < N ≤ 100 é um número perfeito.

Definição: ​ Um número natural é chamado de perfeito se a soma de todos os seus divisores naturais próprios (excluindo ele mesmo) é igual ao próprio número. Exemplo, o número 28 é perfeito, pois 28=1+2+4+7+14. ​

Entrada
```
N
```
Saída

Caso o número informado seja perfeito, imprima:
```
True
```
caso contrário, imprima:
```
False
```

````python
n = int(input())
soma = 0
for i in range(1, n):
  if n % i == 0:
    soma += i
print(soma == n)
````

## Problema 7
Faça um programa que verifica se um número inteiro positivo 0< N< 10⁹, na base 10, é um número de Armstrong.


Entrada:
```
N
```
Saída:

Imprima True ou False para respostas verdadeiras e falsas, respectivamente.
```
True | False (valor de uma variavel booleana)
```
Captura de tela 2021-10-01 180649.jpg

````python
n = input()
num = [float(a) for a in n]
tam = len(num)
armstrong = 0
for i in range(0, tam):
  armstrong += num[i]**tam

print(armstrong == float(n))
  
````

##Problema 8
Dados o centro (x,y,z) e o raio R de uma esfera, calcule a quantidade N de coordenadas inteiras (xi, yi, zi) - onde os valores de xi, yi e zi são números inteiros - que ocorrem no interior (incluindo superfície) da esfera.

Exemplo em 2D:

Captura de tela 2021-10-03 033507.jpg

 

Nesse caso existem 5 pontos no interior do círculo menor e 13 no círculo maior.

Entrada:
```
X Y Z R
```
Saída:
```
N
```
Sample Input 1:
```
0 0 0 1
```
Sample Output 1:
```
7
```

````python
x,y,z,r = [int(num) for num in input().split()]
cont = 0
for xi in range(x-r,x+r+1):
  for yi in range(y-r,y+r+1):
    for zi in range(z-r,z+r+1):
      if (((xi - x)**2) + ((yi - y)**2) + ((zi - z)**2) <= (r**2)):
        cont += 1
print(cont)
````

## Problema 9
Escreva um programa que leia um inteiro n seguido por uma sequência de n + 1 linhas. Destas n + 1 linhas, a última é uma string contendo caracteres cujas ocorrências nas n linhas anteriores devem ser contadas e impressas individualmente na saída do programa.

Entrada

Um número inteiro positivo n, seguido por  n + 1 strings. 

Saída

k linhas, onde k corresponde ao número de caracteres na última string da entrada. Cada linha deve conter o caractere seguido por um espaço em branco, o sinal de igual (=), outro espaço em branco e um número inteiro que corresponda ao número de vezes que o caractere foi encontrado nas nn linhas anteriores.

Confira abaixo alguns exemplos.

Sample Input 1:
```
1
Minha casa
an
```
Sample Output 1:
```
a = 3
n = 1
```
Sample Input 2:
```
2
Meu computador
Minha vida
Mabc
```
Sample Output 2:
```
M = 2
a = 3
b = 0
c = 1
```

````python
n = int(input())
string = ""
for contador in range(n):
  string += input()
busca = input()
for letra in busca:
  qntd = string.count(letra)
  print("%c = %d" % (letra, qntd))
````

Escreva um programa que

peça uma nota entre 0 e 10; 
verifica se a nota inserida está entre 0 e 10;
mostra a mensagem "nota válida", caso o valor inserido seja válido, ou continua pedindo uma nota, caso a nota inserida não seja válida.
Entrada

Valores de notas. Por exemplo
```
11
-5
7
```
Saída

Primeiro seu programa deve imprimir a mensagem 
```
Informe uma nota entre 0 e 10!
```
 depois deve exibir
```
Nota válida!!!
```
caso a nota inserida seja válida, ou 
```
Nota inválida, digite uma nota entre 0 e 10!
```

Confira abaixo alguns casos para um melhor entendimento da entrada e saída do programa!

````python
print("Informe uma nota entre 0 e 10!")
nota = float(input())
while nota < 0 or nota > 10:
  print("Nota inválida, digite uma nota entre 0 e 10!")
  nota = float(input())
print("Nota válida!!!")
````
