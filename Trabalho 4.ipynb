##Problema 1
Faça um programa que leia um vetor de nn posições com valores entre 0 e 20 e encontre:

* A soma dos valores armazenados no vetor;
* O número de ocorrências do valor vn;
* O maior valor armazenado no vetor;
* As posições onde aparecem o maior valor encontrado no item 3. – note que aqui o programa possivelmente imprimirá mais de uma posição.
Entrada
```
v1 v2 v3 ... vn
```
Saída
```
soma
numero de ocorrencias do valor vn
maior valor
indice1 indice2 ... indicek
```

````python
a = [int(i) for i in input().split()]
soma= 0
vn = 0
e = ''
#soma
vn = a.count(a[-1])
for i in range(len(a)):
  soma += a[i]
  if a[i] == max(a):
    e += str(i) + " "

print(soma)
print(vn)       
print(max(a))
print(e)
````

##Problema 2
Faça um programa que preenche um vetor com valores inteiros, fornecidos um por linha até que seja dado um valor negativo (o valor negativo não deve ser inserido no vetor). Depois imprima:

* A quantidade de valores maiores do que 5;
* A soma dos valores pares que foram armazenados no vetor;
* A soma dos valores ímpares que foram armazenados no vetor,
* A quantidade total de valores armazenados no vetor;
* O vetor, em uma linha;

Entrada
```
v1
v2
...
vn
-1
``` 
Saída
```
qtd > 5
soma dos pares
soma dos impares
quantidade de valores
v1 v2 ... vn
```

````python
vetor = []
entrada = 1
somapar = 0
somaimpar = 0
maiorquecinco = 0

while (entrada >= 0):
  entrada = int(input())
  vetor.append(entrada)

vetor.pop(-1)
qtd = len(vetor)

for item in vetor:
  if item % 2:
    somaimpar += item
  else:
    somapar += item
  if item > 5:
    maiorquecinco += 1

print(maiorquecinco)
print(somapar)
print(somaimpar)
print(qtd)
print(" ".join([str(item) for item in vetor]))
````

##Problema 3
Escreva um programa que receba um passo de tempo $\Delta t$, um numero  $M$  de passos de simulação e uma lista com nn partículas com posição inicial e velocidade. Considerando as velocidades constantes e ignorando colisões, o seu programa deve calcular a posição final de cada partícula utilizando o método de integração explícito de Euler:

$\mathbf{p}_i(t+1) = \mathbf{p}_i(t) + \mathbf{v}_i \Delta t$

Onde $\mathbf{p}_i(t) $ representa a posição $(x, y, z)$ da partícula $i$ no instante $t$, e $\mathbf{v}_i $ representa a sua velocidade $(vx, vy, vz)$.

Entrada
```
dt M n
x1 y1 z1 vx1 vy1 vz1
x2 y2 z2 vx2 vy2 vz2
…
xn yn zn vxn vyn vzn
```
Saída
```
x1 y1 z1
x2 y2 z2
…
xn yn zn
```

````python
entrada = [float(num) for num in input().split()]
data = ["x", "y", "z", "vx", "vy", "vz"]

def posicaoFinalDaParticulaR3(x, y, z, vx, vy, vz, dt, M):
  pos = {"x": x,"y": y, "z": z}
  for i in range(int(M)):
    pos["x"] += vx*dt
    pos["y"] += vy*dt
    pos["z"] += vz*dt
  return [pos["x"], pos["y"], pos["z"]]

for i in range(int(entrada[2])):
  particula = dict(zip(data, [float(num) for num in input().split()]))
  posfinal = posicaoFinalDaParticulaR3(*[item for item in particula.values()], entrada[0], entrada[1])
  print("{:.2f} {:.2f} {:.2f}".format(posfinal[0],posfinal[1],posfinal[2]))
````

##Problema 7
Dados os n vértices em sentido anti-horário de um polígono convexo $P$ e $m$ pontos $(x_i,y_i)$. Escreva um programa que verifica se cada ponto $(x_i,y_i)$ se encontra no interior de $P$ (incluindo as arestas) ou não.

Entrada

```
Px1 Py1 Px2 Py2 Px3 Py3 ... Pxn Pyn

x1 y1 x2 y2 x3 y3 ... xm ym
```
Saída

$m$ linhas com S ou N (uma linha para cada ponto).
```
S (se sim) / N (se nao)
```

````python
poligono = [float(num) for num in input().split()]
pontos = [float(num) for num in input().split()]

coord_pontos = []

vertices_poligono = poligono.copy()
pop_poligono = poligono.copy()
vertices = []
x_poligono = []
y_poligono = []

while len(pop_poligono) > 0:
  x_poligono.append(pop_poligono[0])
  pop_poligono.pop(0)
  y_poligono.append(pop_poligono[0])
  pop_poligono.pop(0)

contador_vertices = 0 
lista_suporte_vertices = []
for item in poligono:
  lista_suporte_vertices += [item]
  contador_vertices += 1
  if contador_vertices == 2:
    contador_vertices = 0
    vertices += [lista_suporte_vertices]
    lista_suporte_vertices = []

contador_pontos = 0 
lista_suporte_pontos = []
for item in pontos:
  lista_suporte_pontos += [item]
  contador_pontos += 1
  if contador_pontos == 2:
    contador_pontos = 0
    coord_pontos += [lista_suporte_pontos]
    lista_suporte_pontos = []


todos_os_pontos_do_poligono = []
for i in range(0,len(vertices)):
  p1 = vertices[i]
  p2 = vertices[(i+1)%(len(vertices))]
  if (p2[0] - p1[0]) == 0:
    menor_y = min(p1[1], p2[1])
    maior_y = max(p1[1], p2[1])
    for y in range(int(menor_y*100), int(maior_y*100)+1):
      y /= 100
      todos_os_pontos_do_poligono += [[p1[0], y]]
    todos_os_pontos_do_poligono += ["FIM DA RETA"]
  else:
    coef_ang = (p2[1] - p1[1]) / (p2[0] - p1[0])
    coef_lin = p1[1] - coef_ang * p1[0]
    menor_x = min(p1[0], p2[0])
    maior_x = max(p1[0], p2[0])
    for x in range(int(menor_x*1000), int(maior_x*1000)+1):
      x /= 1000
      y = coef_ang * x + coef_lin
      todos_os_pontos_do_poligono += [[float(num) for num in ("%.2f %.2f" % (x, y)).split()]]
    todos_os_pontos_do_poligono += ["FIM DA RETA"]
    


x_max = max(x_poligono)
x_min = min(x_poligono)

y_max = max(y_poligono)
y_min = min(y_poligono)

for ponto in coord_pontos:
  xp, yp = ponto[0], ponto[1]
  pontos_reta_span = []
  intersec = 0
  for x in range(int(xp*100), int((xp+5)*100)):
      x /= 100
      pontos_reta_span += [[x, yp]]
  i = 0
  while i < len(pontos_reta_span):
    if pontos_reta_span[i] in todos_os_pontos_do_poligono:
      intersec += 1
      i += 10
    i += 1

  if xp > x_max or xp < x_min or yp > y_max or yp < y_min:
    print("N")
  elif ponto in todos_os_pontos_do_poligono:
    print("S")
  elif intersec % 2:
    print("S")
  else: 
    print("N")

# 1 0 0 1 -1 0 0 -1
# 0 0 1.5 1.5 1 0 0.5 -0.2
````

##Problema 8
Faça um programa que leia dois conjuntos de inteiros e mostre a interseção entre eles.



####Entrada

4 linhas com esses dados:
```
tamanho do conjunto A
elementos do conjunto A
tamanho do conjunto B
elementos do conjunto B
```
####Saída

Mostrar a interseção entre os conjuntos A e B:
```
A ∩ B
```
ou, caso a interseção seja vazia, mostrar o caractere '*':
```
*
```

````python
tamA = int(input())
A = [int(num) for num in input().split()]
tamB = int(input())
B = [int(num) for num in input().split()]
intersec = []

for itemA in A:
    for itemB in B:
        if itemA == itemB:
            intersec.append(itemA)
            
if len(intersec) > 0:
    for item in intersec:
        print(item, end=" ")
else:
    print("*")
````

##Problema 10
Um grupo de amigos se reuniu para brincar de loteria. Cada amigo podia apostar em até 10 números diferentes (entre 1 e 100). Depois de todos apostarem, eram sorteados os números. Com base nas apostas e nos números sorteados, o objetivo era calcular o número de acertos de cada um e imprimir o resultado ordenado de forma crescente pelo número de acertos. Havendo empate, a ordem era definida pelo nome.

####Entrada

O programa deve ler um número qualquer de apostas, sendo uma por linha. Cada linha é formada pelo primeiro nome do apostador, de um espaço em branco e de uma sequência de números (entre 1 e 100) separados entre si por um espaço em branco. Chega-se ao fim da leitura das apostas quando se encontra uma linha com a string FIM. Depois de concluída a leitura das apostas, deve ser lido o resultado do sorteio, que pode ter de 1 a 10 números (com valores entre 1 e 100) separados entre si por um hifens.

####Saída

A saída consiste de uma relação de apostadores e seus acertos. A relação deve ser ordenada de forma crescente pelo número de acertos. Em caso de empate, define-se a ordem pelo nome dos apostadores. Separe o nome do apostador e a quantidade de acertos usando o caractere '\t'.

Confira abaixo alguns exemplos.

Sample Input 1:
```
Afonso 4 6 9 7 10 20 8 3
Samuel 1 8 7 10 20 30 5
Filomen 2 5 7 10 19 40
FIM
8-12-30-1-3
```
Sample Output 1:
```
Filomen	0
Afonso	2
Samuel	3
```

````python
entrada = ""
apostas = []
while 1:
  entrada = input()
  if entrada == "FIM":
    break
  lista_entrada = entrada.split()
  apostas.append([lista_entrada[0]] + [lista_entrada[1:]])

buscas = input().split("-")
nomes = [nome[0] for nome in apostas]

qtd = []
for apostador in apostas:
  acertos = []
  for chute in apostador[1]:
    if chute in buscas:
      acertos.append(chute)
  qtd.append(len(acertos))

nomes_e_acertos = list(zip(qtd, nomes))
nomes_e_acertos.sort()

for item in nomes_e_acertos:
  print(str(item[1]) + "\t" + str(item[0]))
````
