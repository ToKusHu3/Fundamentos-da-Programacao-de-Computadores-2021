##Problema 1
Em breve acontecerá uma grande partida de futebol entre Flamengo e Palmeiras valendo o título de campeão da Copa Libertadores da América.

Por conta dos diferentes tipos de ingresso disponíveis e políticas de cobrança, o dinheiro total arrecadado em reais é calculado pela seguinte fórmula:

$f(x) = \left\{\begin{matrix} -100000 & \text{ se } & x = 0\\ x + f(x-100) & \text{se} & x \text{ é par} \\ V + f(x-100) & \text{se} & x \text{ é ímpar} & \end{matrix}\right.$
​
​
 

Em que $V$ representa o valor de um ingresso básico em reais (RS) e $x$ representa a quantidade de pessoas que pagaram ingresso.

Sua tarefa é fazer um programa que dado o valor do ingresso básico calcule a arrecadação total da partida.

**Entrada**
```
Quantidade de pessoas que pagaram ingresso e valor do ingresso básico.
```
**Saída**
```
Valor total da arrecadação.
```
**Sample Input 1:**
```
0 200
```
**Sample Output 1:**
```
R$ -100.000,00
```
**Sample Input 2:**
```
50000 300
```
**Sample Output 2:**
```
R$ 12.425.000,00
```

````python
def valor(x, V):
  total = 0
  if x == 0:
    total += -100000
    return total
  if x%2 == 0:
    total += x + valor(x-100, V)
  if x%2 == 1:
    total += V + valor(x-100, V)
  if x < 0: return total
  return total

entrada = [float(num) for num in input().split()]
retorno = "%.2f" % valor(entrada[0], entrada[1])
final = ""
for i in range(len(retorno) - 3):
  if len(retorno) >= 3:
    if i%3 == 0 and not i == 0:
      if retorno[-4 - i].isnumeric():
        final = "." + final
    final = retorno[-4 - i] + final
    
  
print("R$ " + final + ",00")
````

````python
def soma_lista_rec(L):
  soma = 0
  if isinstance(L, list):
    for item in L:
      soma += soma_lista_rec(item)
  else:
    soma += L  
  return soma
````

##Problema 4
Um dos motivos para aprender recursão é a facilidade de gerar combinações.

Sua missão agora será a de gerar todas as combinações de N lançamentos de moedas e saber quantas dessas têm uma diferença D entre a quantidade de aparições de caras (K)  e coroas (C).

Por exemplo, no lançamento de duas moedas todas as possibilidades são (C,C) , (C,K) , (K,C) e (K,K). Com uma diferença  0 entre quantidades de caras e coroas temos um total de 2 possibilidades, que são (C,K) e (K,C).

Outro exemplo, se fossem 3 moedas, todas as possibilidades seriam (C,C,C), (C,C,K), (C,K,C), (C,K,K), (K,C,C), (K,C,K), (K,K,C) e (K,K,K). No qual em 6 casos ocorre uma diferença 1 entre a quantidade de caras e coroas.

**Entrada**

Dois inteiros **N** e **D** representando a quantidade de moedas e a diferença procurada.

**Saída**

Um inteiro indicando a quantidade de possibilidades que satisfazem a diferença procurada.

**Sample Input 1:**
```
3 1
```
**Sample Output 1:**
```
6
```

````python
from copy import copy

N, D = [int(num) for num in input().split()]

def getPos(lista, N):
  for i in range(N):
      if lista[-1-i] == "C":
        return i

def addPiece(N, tudo = [], lista = []):
  for i in range(2**N-1):
      if lista == []:
        lista = ['C']*N
        tudo += [copy(lista)]

      if lista.count("K") == N:
        return tudo
      else:
        pos = getPos(lista, N)
        lista[-1-pos] = 'K'
        for i in range(pos):
          lista[-1-i] = 'C'
        tudo += [copy(lista)]
  return tudo
 
def encontrarD(perms, D):
  cont = 0
  for perm in perms:
    if D == abs(perm.count("C") - perm.count("K")):
      cont += 1
  return cont

if N%2 == 0 and D%2 == 1:
    print(0)
elif N%2 == 1 and D%2 == 0:
    print(0)
else:
    tudo = addPiece(N)
    print(encontrarD(tudo, D))
````

O primeiro nível de indentação é o nível zero;
O caractere de tabulação (\t) deve ser usado para espaçar cada nível de indentação: nível 0 = 0 tab, nível 1 = 1 tab, nível n = n tabs .
Cabeçalho da função:

def print_dict_rec(D, level=0):

Não se preocupe com a chamada das funções, ou impressão dos resultados, isso será feito automaticamente.  
Escreva apenas as funções solicitadas.

**Sample Input 1:**
```
{'a': 1, 'b': {'c': 2, 'd': {'e': 3, 'f': {'g': 4, 'h': {}}}}}
```
**Sample Output 1:**
```
a: 1
b: {
	c: 2
	d: {
		e: 3
		f: {
			g: 4
			h: {
			}
```

````python
def print_dict_rec(D,level=0):
  for key,value in D.items():
    if isinstance(value, dict):
      print(level*"\t" + key + ": {")
      print_dict_rec(value, level+1)
      print(level*"\t" + "}")
    else:
      print(level*"\t" + key + ": " + str(value))
  
  return ""
````

##Problema 6
Eis que o tabuleiro de rainhas esta de volta! Escreva um código que encontre de quantas maneiras diferentes e possível organizar N rainhas em um tabuleiro $N \times N$ sem que nenhuma rainha ameace a outra. Lembrando que uma rainha caminha na horizontal, vertical ou diagonal. 

**Sample Input 1:**
```
8
```
**Sample Output 1:**
```
92
```

````python
class NDamas:
    def __init__(self, tamanho):
        self.tamanho = tamanho
        self.solução = 0
        self.resolver()

    def resolver(self):
        posições = [-1] * self.tamanho
        self.coloca_dama(posições, 0)
        print(self.solução)

    def coloca_dama(self, posições, fileira_alvo):
        if fileira_alvo == self.tamanho:
            self.solução += 1
        else:
            for coluna in range(self.tamanho):
                if self.checar_posição(posições, fileira_alvo, coluna):
                    posições[fileira_alvo] = coluna
                    self.coloca_dama(posições, fileira_alvo + 1)

    def checar_posição(self, posições, fileiras_ocupadas, coluna):
        for i in range(fileiras_ocupadas):
            if posições[i] == coluna or \
                posições[i] - i == coluna - fileiras_ocupadas or \
                posições[i] + i == coluna + fileiras_ocupadas:
                return False
        return True

def main(n):
    NDamas(n)

n = int(input())
main(n)
````

Dica: Esta não precisa ser a função recursiva da sua solução, esta função pode chamar a sua função recursiva la dentro. OBS: Considere que o índice do primeiro elemento dessa sequência é 1.

**Entrada**

Um valor inteiro, o  índice do elemento na sequência

**Saída**

Um valor inteiro, o valor dentro da série. Para índices fora do range, ou seja, ou seja menores ou iguais a zero, a saída será SEM RESPOSTA.

**Sample Input:**
```
4
```
**Sample Output:**
```
12
```

````python
def SerieNeymarzinho(i, cont = 0):
  valor = 0
  if i <= 0:
    return print("SEM RESPOSTA")
  elif i == 1:
    return 3
  else:
    if cont == 0:
      cont = 1
      troca = 4
    else:
      cont = 0
      troca = 1
    valor = troca + SerieNeymarzinho(i-1, cont)
    return valor
````
