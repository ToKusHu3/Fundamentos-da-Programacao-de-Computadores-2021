##Problema 1
Dado um array de números inteiros, seu trabalho deve ser encontrar quais elementos do array original continuam na mesma posição depois que o array for ordenado e qual é sua posição. Considere que a posição do primeiro elemento é 1 ao invés de 0.

Exemplo:

* Array original: 3, 2, 4, 10, 9, 11.

* Array ordenado: 2, 3, 4, 9, 10, 11.

* Número 4, posição 3

* Número 11, posição 6


**Entrada**

Um número N representando a quantidade de elementos do vetor e um vetor de números inteiros. (0 < n < 1000)


**Saída**

A quantidade de elementos que manteve sua posição depois da ordenação do vetor e esses elementos junto com sua posição no vetor.

**Sample Input 1:**
```
6
3 2 4 10 9 11
```
**Sample Output 1:**
```
2
4 3
11 6
```

````python
from copy import copy
n = int(input())

array = [int(num) for num in input().split()]
sortedarray = sorted(array.copy())  
coincidence = []
cont = 0

for i in range(n):
    if array[i] == sortedarray[i]:
        cont+=1
        coincidence.append((array[i], i+1))

print(cont)
for i in range(cont):
  print("%i %i" %(coincidence[i][0], coincidence[i][1]))
````

##Problema 2
Escreva um programa que leia uma lista de números inteiros e imprima-os na saída de maneira ORDENADA da seguinte forma:

Se o primeiro número lido for par ordem será crescente.
Caso contrário, será decrescente.
**Entrada**

Uma linha com uma sequência de números inteiros que devem ser ordenados de acordo com os critérios mencionados acima.

**Saída**

Os números ordenados ou a mensagem de erro
```
Lista inválida!
```
se qualquer um dos valores da lista for negativo.

**Sample Input 1:**
```
-1 1 56 66 75 85 93 95
```
**Sample Output 1:**
```
Lista inválida!
```

````python
def getLista():
    listafinal = []
    lista = [int(num) for num in input().split()]
    n = len(lista)
    lista.sort()
    for item in lista:
        if item < 0:
            print("Lista inválida!")
            return
    if lista[0]%2 == 0:
        print(" ".join([str(item) for item in sorted(lista)]))
    else:
        for i in range(n-1,-1, -1):
            listafinal.append(lista[i])
        print(" ".join([str(item) for item in listafinal]))
    
getLista()
````

##Problema 5
Escreva um programa que receba duas listas de inteiros $[x_1,x_2,..., x_m]$ e $[y_y,y_2,..., y_n]$
 ordenadas e produza uma outra lista $[z_1,z_2,..., z_{m+n}]$ordenada que contenha o resultado da intercalação das duas listas dadas. 

Caso uma das listas de entrada não esteja ordenada, o programa deve retornar -1 como código de erro. 

**Entrada**

Duas linhas com duas listas de inteiros.

**Saída**

Uma linha com os elementos da lista $[z_1,z_2,..., z_{m+n}]$. Caso uma das listas de entrada não esteja ordenada, deve-se imprimir -1.

**Sample Input 1:**
```
24 29 21 25 80
13 18 24 31 32 33 40 41 54 56 61 70 71 79 79 83 85
```

```

```

````python
lista1 = [int(num) for num in input().split()]
lista2 = [int(num) for num in input().split()]
incomp = False
lista1sort = sorted(lista1)
lista2sort = sorted(lista2)

for i1, i2 in zip(lista1, lista1sort):
    if not i1 == i2:
        incomp = True
        
for i1, i2 in zip(lista2, lista2sort):
    if not i1 == i2:
        incomp = True       
        
if incomp:
    print(-1)
else:
    print(*sorted(lista1 + lista2))
````

##Problema 8
Faca um programa que imprima a tabela de classificação de um campeonato de futebol a partir dos resultados dos jogos. A pontuação se dá da seguinte maneira:

* O time ganha 3 pontos para cada vitória;
* 1 ponto para cada empate;
* 0 pontos para cada derrota.

**Entrada**

A primeira linha de entrada contém um número $N$, que corresponde ao número de jogos que já ocorreram. Cada uma das $N$ linhas que se segue, contêm uma tupla representando o resultado de cada jogo no seguinte formato:
```
(TIME_A, GOLS_A, GOLS_B, TIME_B)
```
**Saída**
```
Lista de times, um por linha.
```
**Sample Input 1:**
```
3
('time_a', 1, 0, 'time_b')
('time_c', 1, 0, 'time_b')
('time_a', 1, 3, 'time_c')
```
**Sample Output 1:**
```
time_c
time_a
time_b
```

````python
times = {}
n = int(input())
jogos = []
for i in range(n):
  jogos.append(eval(input()))

for jogo in jogos:
  if jogo[0] not in times:
    times[jogo[0]] = 0
  if jogo[3] not in times:
    times[jogo[3]] = 0
  if jogo[1] > jogo[2]:
    times[jogo[0]] += 3
  elif jogo[1] < jogo[2]:
    times[jogo[3]] += 3
  else:
    times[jogo[0]] += 1
    times[jogo[3]] += 1
results = sorted(times.items(), key=lambda item: item[1], reverse=True)

for item in results:
  print(item[0])
````

##Problema 9
Em uma certa empresa que trabalha com mapas existe um sistema de marcação que é feito no sentido horário. Um de seus funcionários descobriu que teria uma maneira mais eficaz de fazer isso e seria mudando a ordem dos pontos. Então ele contatou o setor de tecnologia e você ficou encarregado de criar um programa que inverta esses pontos.

Você receberá um conjunto de pontos e deverá imprimi-los no sentido horário, começando do ponto mais a direita (maior valor da coordenada X).

A figura abaixo mostra como seria a ordem de impressão para os pontos dados no exemplo de entrada e saída nesse exercício.

image.png

 

Dica: Analise os ângulos entre os vetores formados pelo centroide dos pontos e os próprios pontos.

image.png


**Entrada**

Consiste de uma sequência de n pontos, um ponto por cada linha. Os pontos tem o seguinte formato: x, y (onde x e y são números com uma quantidade de casas decimais que varia de 0 a 15, positivos ou negativos). A leitura dos pontos é encerrada pelo caractere '*'.

**Saída**

Seu programa deve imprimir os pontos ordenados de acordo com o sentido horário. O primeiro ponto é o ponto mais a direita (maior x). Você também deve imprimir os pontos com a mesma quantidade de casas decimais da entrada.

**Sample Input:**
```
-9.747971411763078,-36.640434265136720
-9.75981400716881,-36.650047302246094
-9.756768808542352,-36.680946350097656
-9.747294679319221,-36.669273376464844
-9.745264473749616,-36.662750244140625

```
**Sample Output:**
```
-9.745264473749616,-36.662750244140625
-9.747294679319221,-36.669273376464844
-9.756768808542352,-36.680946350097656
-9.75981400716881,-36.650047302246094
-9.747971411763078,-36.640434265136720
```

````python
import math

def getPoints():
  pontos = []
  entrada = input().split(',')
  while not entrada == ['*']:
    pontos.append(entrada)
    entrada = input().split(',')
  return pontos

def getXY(points):
  n = len(points)
  xcm = 0
  ycm = 0
  for ponto in points:
    xcm += ponto[0]
    ycm += ponto[1]
  xcm /= n
  ycm /= n
  points.append([xcm, ycm])

def getVectors(points):
  vetores = []
  for i in range(len(points)-1):
    vetores.append([points[i][0] - points[-1][0], points[i][1] - points[-1][1]])
  maxindex = 0
  for ponto in points:
    if ponto[0] > points[maxindex][0]:
      maxindex = points.index(ponto)
  return maxindex, vetores

def getAngle(vec1, vec2):
  cosseno = (vec1[0]*vec2[0] + vec1[1]*vec2[1])/(math.sqrt(vec1[0]**2 + vec1[1]**2)*math.sqrt(vec2[0]**2 + vec2[1]**2))
  dot = vec1[0]*-vec2[1] + vec1[1]*vec2[0]
  angulo = math.degrees(math.acos(round(cosseno, 3)))
  if dot < 0:
    return -angulo + 360
  return angulo

pontosstr = getPoints()
pontos = []
for par in pontosstr:
  pontos.append([float(par[0]), float(par[1])])
getXY(pontos)
idxmaiorx, vetores = getVectors(pontos)
pontossemcm = pontos[:-1]

relation = {}

for i in range(0, len(vetores)):
  relation[i] = getAngle(vetores[idxmaiorx], vetores[i])

orderedrelation = dict(sorted(relation.items(), key=lambda item: item[1]))

for i in orderedrelation.keys():
  print(pontosstr[i][0]+','+pontosstr[i][1])
````
